```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

cookies = {
    'cookie_name': 'cookie_value'
}
```

```python
url= "https://www.espn.com.mx/futbol-americano/nfl/estadisticas/equipo"
```

```python
r = requests.get(url, cookies=cookies, headers=headers)
html_contents = r.text
```

```python
html_soup = BeautifulSoup(html_contents, 'html.parser')
```

```python
tabla_puntaje = html_soup.find_all('table', class_='Table Table--align-right') 
```

```python
columnas =  ['GP','YDS','YDS/G','YDS','YDS/G',
             'YDS','YDS/G','PTS','PTS/G',]
```

```python
inf = pd.DataFrame(columns = columnas)
```

```python
for table in tabla_puntaje:
    for row in table.find_all('tr')[2:]:
        values = []
        for col in row.find_all(['th','td']):
            values.append(col.text)
        if len(values) == len(columnas):
            inf.loc[len(inf)] = values
```
